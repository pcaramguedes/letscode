```python
#Faça um script que informe o fatorial de um número:

#Exemplo
#input: 5
#output: 120
#Exemplo
#input: 3
#output: 6

# Não temos fatorial de numeros decimais, por isso a variavel foi setada para receber apenas inteiros

while True:
    try:
        fatorial = int(input('Entre com um numero inteiro: '))
        x = fatorial
        
        if fatorial == 1 or fatorial == 0:
            print(f'O fatorial de {fatorial} é 1')
            break
        elif fatorial < 0:
            print('Não temos fatorial de números negativos !!!')
            break
           
    except ValueError:
        print('Favor entrar somente com números inteiros !!')
    
    else:
        for num in range(1,fatorial):
            fatorial *= (x-num)
         
        print('='*32)   
        print(f'O fatorial de {x} é {fatorial}')
        break

```

```python
#Informe, de forma decrescente todos os pares entre N (número fornecido pelo usuário) e -N.

#Exemplo
#input: 5
#output: 4
#        2
#        0
#        -2
#        -4
#Exemplo
#input: 3
#output: 2
#        0
#       -2


try:
    numero = int(input('Digite um numero qualquer: '))

except ValueError:
    print('Digite apenas números inteiros !!!')

else:
    if numero >0:
        for x in range(numero,-numero,-1):
            if x % 2 == 0:
                print(x)
    else:
        print(f'{numero} digitado não pode ser negativo')
    
    
```

```python
#Faça um script que reproduza o padrão usando "*" a seguir de acordo com o número de linhas desejadas pelo usuário.
#Dica: print(5 * '@')

#Exemplo 1
#input: 3
#output: *
#        **
#        ***
#Exemplo 2
#input: 5
#output: *
#        **
#        ***
#        ****
#        *****

try:
    linhas = int(input('Entre com o número de linhas: '))

except ValueError:
    print('\nDigite apenas números inteiros')

else:
    for num in range(linhas+1):
        print('*'*num)
```
